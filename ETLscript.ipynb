```python
# import dependancies

import pandas as pd
```

```python
# read in the data
df = pd.read_csv('C:/Users/tgruh/Desktop/nirvana_performances.csv')
df.head()
```

```python
#remove the recording_surface column
df.drop('recording_surfaced', axis=1, inplace=True)
df.head()
```

```python
#split the date column into 3 columns for year, month, and day
# handle null values by filling them with na
df['month'] = df['date'].str.split('/').str[0]
df['day'] = df['date'].str.split('/').str[1]
df['year'] = df['date'].str.split('/').str[2]

#show all rows
df.head(100)
```

```python
#set any values in the month and day columns that are not 
df['month'] = pd.to_numeric(df['month'], errors='coerce')
df['day'] = pd.to_numeric(df['day'], errors='coerce')
df.head()
```

```python
df['day'] = df['day'].astype(float).astype('Int64')
df['month'] = df['month'].astype(float).astype('Int64')
df['year'] = df['year'].astype(float).astype('Int64')
df.head()
```

```python
#convert years from 2 digit to 4 digit
df.loc[df['year'] < 100, 'year'] += 1900
df.head(100)
```

```python
#delet the date column
df.drop('date', axis=1, inplace=True)
df.head()
```

```python
#export the data to a csv file called nirvana_performances_cleaned.csv
df.to_csv('C:/Users/tgruh/Desktop/nirvana_performances_cleaned.csv', index=False)
```
