```python
import http.client
import json
from pprint import pprint

def get_artist_events(api_key, artist_name):
    conn = http.client.HTTPSConnection("www.jambase.com")
    headers = {'Accept': "application/json"}

    conn.request("GET", f"/jb-api/v1/events?apikey={api_key}&artistName={artist_name}", headers=headers)

    res = conn.getresponse()
    data = res.read()

    decoded_data = data.decode("utf-8")
    json_data = json.loads(decoded_data)

    return json_data

# Replace 'YOUR_API_KEY' with your actual JamBase API key
api_key = "bdf2d4f5-f2b8-4dba-91bf-0c2c5d8fadf3"
artist_name = "Tool"
events_json = get_artist_events(api_key, artist_name)

pprint(events_json)
```
